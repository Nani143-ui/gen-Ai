```python
import torch
from diffusers import PixArtAlphaPipeline
```

```python

# Model ID
model_id = "PixArt-alpha/PixArt-XL-2-512x512"

# Select dtype and device
dtype = torch.float16 if torch.cuda.is_available() else torch.float32
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load pipeline
pipe = PixArtAlphaPipeline.from_pretrained(
    model_id,
    torch_dtype=dtype,
)

# Move to device
pipe = pipe.to(device)

# Optional: enable faster attention if on GPU with limited VRAM
if device == "cuda":
    pipe.enable_model_cpu_offload()

# Enable VAE slicing for memory efficiency
pipe.enable_vae_slicing()

# Test prompt
prompt = "A futuristic cityscape with flying cars at sunset"
image = pipe(prompt).images[0]

# Save the result
image.save("pixart_output.png")
```

```python
# Define prompt and negative prompt properly
prompt = (
    "Ultra-realistic close-up portrait of a young adult human, neutral expression, "
    "natural soft lighting, 85mm lens bokeh, high detail skin, photographic quality"
)

negative_prompt = (
    "blurry, low quality, extra fingers, extra eyes, deformed, text, watermark, logo"
)

# Generate the image
image = pipe(
    prompt=prompt,
    negative_prompt=negative_prompt,
).images[0]

# Save image
image.save("portrait_output.png")
```

```python


# Set generator for reproducibility
generator = torch.Generator(device=device).manual_seed(42)

# Generate the image
image = pipe(
    prompt=prompt,
    negative_prompt=negative_prompt,
    num_inference_steps=24,   # increase for slightly higher quality
    guidance_scale=4.5,       # 3.5–6 is typical
    height=512,
    width=512,
    generator=generator       # <-- needs '='
).images[0]

# Save the result
image.save("face.png")
print("Saved to face.png")
```
